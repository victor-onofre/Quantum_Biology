```python
def Calcular_elemento(SigmasBaseRho,evals,label1,label2):
    g=[]# Lista donde se guardara los elementos para calcular el elmento de matriz W[label1.label2]  
    for i in range(0,10):
        for j in range(0,10):
            if i==j:
                g.append(0)
            elif (evals[i]+evals[j]) < 0.0000000000000000001:
                g.append(0) # Si el denominador de la formula para obtener la elemento W[label1.label2] es cercana a cero, el elemnento guardado sera cero  
            else: 
                g.append( ((2*(evals[i]*evals[j]))/(evals[i]+evals[j]))*(SigmasBaseRho[label1][i,j]*SigmasBaseRho[label2][j,i] )  )  # Calcula el elemento de matriz W[label1.label2]                                   
    return g
```

## Caso 1 : $H = H_{A}$

Para $ \rho_{steady} = \sum_{i} \lambda_{m} | \psi_{i} > < \psi_{i} |$
\begin{equation}
\begin{aligned}
    F(\rho,H_{A}) & = \frac{1}{2} \sum_{m,n} \frac{(\lambda_{i} - \lambda_{j})^2}{\lambda_{i}+\lambda_{j}} | <i|H_{A} | j > |^2
    \end{aligned}
\end{equation}
se define H como:
\begin{equation}
   \begin{aligned}
    H_{A} = \boldsymbol{n}_{\uparrow} \cdot \boldsymbol{\sigma}_{\uparrow} 
    \end{aligned}
\end{equation}

donde $ \boldsymbol{\sigma}_{\uparrow} = \sigma_{x,\uparrow} \hat{i} + \sigma_{y,\uparrow} \hat{j} + \sigma_{z,\uparrow} \hat{k}$ La informacion cuantica de Fisher se define como

\begin{equation}
Fisher_{A}= \lambda_{min}(W)
\end{equation}
, el eigenvalor m√≠nimo de la matriz $W$ de $3 \times 3$:

\begin{equation}
\begin{aligned}
    (W_{\uparrow,\downarrow})_{mn}  = \frac{1}{2} \sum_{i,j} \frac{(\lambda_{i} - \lambda_{j})^2}{\lambda_{i}+\lambda_{j}} < \Psi_{i} | \sigma_{m,\uparrow} | \Psi_{j} > < \Psi_{j}| \sigma_{n,\uparrow} | \Psi_{i} |>
    \end{aligned}
\end{equation}

## Caso 3 : $H = H_{A} + H_{B}$

Para $ \rho_{steady} = \sum_{i} \lambda_{m} | \psi_{i} > <\psi_{i}|$
\begin{equation}
\begin{aligned}
    F(\rho,H_{A}) & = \frac{1}{2} \sum_{i,j} \frac{(\lambda_{i} - \lambda_{j})^2}{\lambda_{i}+\lambda_{j}} | < i |( H_{A} + H_{B}) | j >|^2 \\
     & =\frac{1}{2} \sum_{i,j} \frac{(\lambda_{i} - \lambda_{j})^2}{\lambda_{i}+\lambda_{j}} \{ | < i | H_{A} | j >|^2 + | < i| H_{B} | j >|^2 +  < i| H_{B} | j> < j| H_{A} | i> +  < i| H_{A} | j> < j | H_{B} | i> \}
    \end{aligned}
\end{equation}
se define H como:
\begin{equation}
   \begin{aligned}
    H_{A} + H_{B} =\boldsymbol{n}_{\uparrow} \cdot \boldsymbol{\sigma}_{\uparrow}  +  \boldsymbol{n}_{\downarrow} \cdot \boldsymbol{\sigma}_{\downarrow} 
    \end{aligned}
\end{equation}

La informacion cuantica de Fisher se define como

\begin{equation}
Fisher_{A}= \lambda_{min}(W)
\end{equation}
, el eigenvalor m√≠nimo de la matriz $W$ de $3 \times 3$:

\begin{equation}
\begin{aligned}
    (W_{\uparrow})_{mn}  & = \frac{1}{2} \sum_{i,j} \frac{(\lambda_{i} - \lambda_{j})^2}{\lambda_{i}+\lambda_{j}} < \Psi_{i}| \sigma_{m} | \Psi_{j}> <\Psi_{j}| \sigma_{n}| \Psi_{i}> 
     + \frac{1}{2} \sum_{i,j} \frac{(\lambda_{i} - \lambda_{j})^2}{\lambda_{i}+\lambda_{j}} < \Psi_{i} | \sigma_{m,} | \Psi_{j} > < \Psi_{j}| \sigma_{n} | \Psi_{i} > \\
    & + \frac{1}{2} \sum_{i,j} \frac{(\lambda_{i} - \lambda_{j})^2}{\lambda_{i}+\lambda_{j}} < \Psi_{i}| \sigma_{m}  |\Psi_{j} > < \Psi_{j}| \sigma_{n} | \Psi_{i} > 
     + \frac{1}{2} \sum_{i,j} \frac{(\lambda_{i} - \lambda_{j})^2}{\lambda_{i}+\lambda_{j}} < \Psi_{i} | \sigma_{m}  |\Psi_{j}> <\Psi_{j}| \sigma_{n} | \Psi_{i} >
    \end{aligned}
\end{equation}

\begin{equation}
\begin{aligned}
    (W_{\uparrow,\downarrow})_{mn}  & = \frac{1}{2} \sum_{i,j} \frac{(\lambda_{i} - \lambda_{j})^2}{\lambda_{i}+\lambda_{j}} \big\{ < \Psi_{i} |  \sigma_{m} | \Psi_{j} > < \Psi_{j} | \sigma_{n} | \Psi_{i} > 
    + < \Psi_{i}| \sigma_{m} | \Psi_{j} > < \Psi_{j}| \sigma_{n} | \Psi_{i} > \\
    & +  < \Psi_{i}| \sigma_{m} | \Psi_{j} > < \Psi_{j}| \sigma_{n}| \Psi_{i}> 
    + < \Psi_{i}| \sigma_{m} |\Psi_{j}> < Psi_{j}| \sigma_{n}|\Psi_{i}> \big\}
    \end{aligned}
\end{equation}


donde $ \sigma_{n} = (\sigma_{1},\sigma_{2},\sigma_{3},\sigma_{4},\sigma_{5},\sigma_{6}) = ( \sigma_{x,\uparrow},  \sigma_{y,\uparrow}, \sigma_{z,\uparrow},\sigma_{x,\downarrow},  \sigma_{y,\downarrow}, \sigma_{z,\downarrow})$
